## Using your store items data:

### 1. Convert date column to datetime format.

```python
# imports
import pandas as pd
from acquire import get_store_data, get_germany
# get store items data
sales = get_store_data()
sales.head()
```

```python
# convert sale_date to datetime
sales.sale_date = pd.to_datetime(sales.sale_date)
# confirm dtype
sales.info()
```

### 2. Plot the distribution of `sale_amount` and `item_price`.

```python
# plot distribution of sale_amount
sales.sale_amount.hist()
```

```python
# plot distribution of item_price
sales.item_price.hist()
```

### 3. Set the index to be the datetime variable.

```python
# set index to dt var, sort index
sales = sales.set_index('sale_date').sort_index()
sales.head()
```

### 4. Add a 'month' and 'day of week' column to your dataframe.

```python
# add month column
sales['month'] = sales.index.month_name()
sales.head()
```

```python
# add day of week column
sales['day_of_week'] = sales.index.day_name()
sales.head()
```

### 5. Add a column to your dataframe, `sales_total`, which is a derived from `sale_amount` (total items) and `item_price`.

```python
# add sales_total column
sales['sales_total'] = sales.sale_amount * sales.item_price
sales.head()
```

### 6. Make sure all the work that you have done above is reproducible. That is, you should put the code above into separate functions and be able to re-run the functions and get the same results.

```python
# combine above into function
def prepare_sales(sales):
    '''
    This function takes in the sales dataframe and returns a prepared version
    of the sales dataframe.
    '''
    sales.sale_date = pd.to_datetime(sales.sale_date)
    sales = sales.set_index('sale_date').sort_index()
    sales['month'] = sales.index.month_name()
    sales['day_of_week'] = sales.index.day_name()
    sales['sales_total'] = sales.sale_amount * sales.item_price
    return sales
# test function
prepare_sales(get_store_data())
```

## Using the OPS data acquired in the Acquire exercises `opsd_germany_daily.csv`, complete the following:

### 1. Convert date column to datetime format.

```python
# get data
germany = get_germany()
germany.head()
```

```python
# convert Date to datetime
germany.Date = pd.to_datetime(germany.Date)
germany.info()
```

### 2. Plot the distribution of each of your variables.

```python
# plot all variable distributions
import matplotlib.pyplot as plt
germany.hist(figsize=(10,6))
plt.tight_layout();
```

### 3. Set the index to be the datetime variable.

```python
# set date as index
germany = germany.set_index('Date').sort_index()
germany.head()
```

### 4. Add a month and a year column to your dataframe.

```python
# add month column
germany['month'] = germany.index.month_name()
germany.head()
```

```python
# add year column
germany['year'] = germany.index.year
germany.head()
```

### 5. Fill any missing values.

```python
# fill missing values with 0
germany = germany.fillna(0)
germany.head()
```

### 6. Make sure all the work that you have done above is reproducible. That is, you should put the code above into separate functions and be able to re-run the functions and get the same results.

```python
# combine above into function
def prepare_germany(germany):
    germany.Date = pd.to_datetime(germany.Date)
    germany = germany.set_index('Date').sort_index()
    germany['month'] = germany.index.month_name()
    germany['year'] = germany.index.year
    germany = germany.fillna(0)
    return germany

# test function
prepare_germany(get_germany())
```
